```python
# Replace income Values ['<=50K', '>50K'] With 0 and 1
df.columns
```

```python
df['income'].unique()
```

```python
df['income'].value_counts()
```

```python
sns.countplot(x='income',data=df)
```

```python
# replace with 0 and 1 using function
def income_data (salary):
  if salary=='<=50k':
    return 0
  else:
    return 1
```

```python
df['encoded_salary']=df['income'].apply(income_data)
```

```python
df.head(1)
```

```python
# replace with 0 and 1
df['income'].replace(['<=50K','>50K'],[0,1],inplace=True)
```

```python
df.head(1)
```

```python
# Which Workclass Getting The Highest Salary?
df.groupby('workclass')['income'].mean().sort_values(ascending=False)
```
